### Defining a NamedColor class 

Let's start with a little bit of a silly example.

```python
class NamedColor():
    name = "Blue"
    rgb = [0,0,255]
```

Let's create an instance of `NamedColor()`

```python
x = NamedColor()
```

Behind the scenes: `__init__` is being called. We can update this ourselves.

```python
x.name
```

```python
x.rgb
```

This example is a little lacking because there's no variation! Maybe we don't want everything to be blue. Idea: Pass arguments that tell us how to set the color.

```python
class NamedColor():
    def __init__(self, name, rgb_tuple):
        self.name = name
        self.rgb = rgb_tuple
```

```python
x = NamedColor()
```

What's happening here is I didn't specify any input arguments. When it says "positional arguments" that means I can just pass the values, without needing keywords

```python
x = NamedColor("Red",[255,0,0])
```

```python
x.name
```

```python
x.rgb
```

Now let's try to update our code so that I don't necessarily need to pass an RGB triple.

```python
from matplotlib.colors import to_rgb
```

```python
to_rgb("Red")
```

Notice, this is normalized different than we're used to. Here's how we can update it:

```python
[round(255*x) for x in to_rgb("Red")]
```

Let's set a default value for `rgb_tuple`

```python
class NamedColor():
    def __init__(self, name, rgb_tuple = None):
        if rgb_tuple is None:
            rgb_tuple = [round(255*x) for x in to_rgb(name)]
        self.name = name
        self.rgb = rgb_tuple
```

```python
z = NamedColor("Orange")
```

```python
z.name
```

```python
z.rgb
```

```python
z #calling __repr__
```

```python
print(z) #calling __str__
```

```python
class NamedColor():
    def __init__(self, name, rgb_tuple = None):
        if rgb_tuple is None:
            rgb_tuple = [round(255*x) for x in to_rgb(name)]
        self.name = name
        self.rgb = rgb_tuple
    
    def __str__(self):
        return f"The color {self.name} with RGB value {tuple(self.rgb)}"
```

```python
z = NamedColor("Green")
```

```python
z
```

```python
print(z)
```
